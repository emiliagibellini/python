# Introducción a Python

## Sentencias

Una sentencia o *statement* es una declaración que hacemos en términos que Python pueda interpretar para ejecutar alguna acción.

Ejemplo:

```python
nombre = 'Emilia'
```

Lo que estamos haciendo es crear *nombre* y asignarle el valor "Emilia". Esto se conoce como una sentencia de asignación. Más adelante veremos otros tipos de sentencias.

Imprimámoslo para ver qué información guardó:

```python
print(nombre)
```

### Sentencias con varios renglones

A veces podemos tener una sentencia larga y podríamos querer escribirla en varios renglones. Podemos hacerlo valiéndonos de "\".

Por ejemplo:

```python
total = 4 + 3 + 2 + \
        5 + 11 \
        + 7

total
```

Otra forma válida también podría ser:

```python
total = (4 + 3 + 2 +
         5 + 11 
         + 7)
```

Notemos que al tener los paréntesis "()" queda implícito que la suma continúa en los renglones de abajo.

## Comentarios

En ocasiones puede ser útil dejar algún comentario junto con nuestro código. Para hacerlo, usamos "#". Todo lo que escribamos después del "#" no se ejecutará.

Por ejemplo:

```python
nombre = "mi_nombre" # Completar con su nombre

print(nombre) # Imprimimos
```

## Palabras clave

Las palabras claves o *keywords* son denominaciones que tienen algún objetivo específico dentro de Python, y por lo tanto no deberíamos utilizarlas para nombrar entidades (clases, variables, funciones, entre  otras). Por ahora no vamos a adentrarnos en qué significa cada una, simplemente las vamos a consultar para tener una idea de cuáles son estas palabras.

```python
help('keywords')
```

Para ampliar información sobre una de ellas podemos hacer esta consulta:

```python
help('None')
```
