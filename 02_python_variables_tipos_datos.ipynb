# Variables

Las variables son objetos que almacenan un valor, que puede ser modificado.

Algunos ejemplos:

```python
a = 'A'
b = 'alfabeto'
c = 24
d = 1e3
e = True

print(a)
print(b)
print(c)
print(d)
print(e)
```

```python
c = 58
print(c)
```

La variable *c* fue modificada.

También es posible crear múltiples variables en un solo paso:

```python
a, b, c, d, e = 'A', 'lfabeto', 24, 1e3, True

print(a)
print(b)
print(c)
print(d)
print(e)
```

Veamos qué pasa si usamos el signo + con dos variables que almacenan cadenas de caracteres.

```python
a + b
```

¿Y si tenemos variables con distintos tipos de datos?

```python
a + d
```

```python
a + e
```

## Tipos de datos

Entre los tipos de datos más elementales encontramos:

- ``str`` - cadena de caracteres
- ``int`` - números enteros
- ``float`` - números con decimales
- ``bool`` - True / False

Es posible conocer el tipo de dato usando la función *type*. Ejemplos:


```python
print(a, type(a))
print(b, type(b))
print(c, type(c))
print(d, type(d))
print(e, type(e))
```
